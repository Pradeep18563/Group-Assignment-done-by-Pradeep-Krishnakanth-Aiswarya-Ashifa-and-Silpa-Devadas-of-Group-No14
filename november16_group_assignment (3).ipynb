# Compare churn count with respect to gender.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_excel(r"C:\Users\HP\OneDrive\Desktop\Telco-Customer-Churn.xlsx")
```

```python
print('Number of Males and Females\n',data['gender'].value_counts())
sns.countplot(data=data,x='gender',hue='Churn')#churn rate gender wise
plt.grid()
```

```python
female_senior=data.groupby(['gender','SeniorCitizen']).size().reset_index().pivot(columns='gender',index='SeniorCitizen',values=0)
female_senior
```

# Comparing 'tenure' with 'Total Charges'

```python
data['TotalCharges']=pd.to_numeric(data.TotalCharges,errors='coerce')
data['TotalCharges'].dtypes
```

```python
x=data['tenure']
y=data['TotalCharges']
plt.figure(figsize=(15,10))
plt.bar(x,y)
plt.title('Graph of Tenure vs Total Charges')
plt.xlabel('Tenure')
plt.ylabel('Total Charges')
plt.show()
```

```python
#Insight: Total Charges increases with increase in Tenure
```
